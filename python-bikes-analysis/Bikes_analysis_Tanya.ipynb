#                                                          Bike Rental Data Analysis
#### Introduction

Bay Wheels is a bike-sharing system that provides an efficient and eco-friendly transportation option across urban areas. With the growing adoption of sustainable travel, understanding user behavior is key to optimizing operations, improving customer experience, and increasing ridership. This analysis explores patterns in bike rentals based on time, weather, and holidays. The goal is to uncover actionable insights that can inform strategic planning. Using a dataset of over 1,300 records, we aim to determine what drives bike usage and how the company can leverage environmental and behavioral patterns to grow.






```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data from Excel file provided on the assignment
file_path = 'bikes_data.xlsx'
df = pd.read_excel(file_path, sheet_name='SU (2 + 3)')
```

```python
# --- Visualization 1: Rental Volume by Hour ---
rentals_by_hour = df.groupby('Hour', observed=True)['RENTALS'].mean().reset_index()
plt.figure(figsize=(10, 5))
sns.lineplot(x='Hour', y='RENTALS', data=rentals_by_hour, marker='o')
plt.title('Average Bike Rentals by Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Average Rentals')
plt.grid(True)
plt.xticks(range(0, 24))
plt.show()

# This visual shows peak rental times around morning and evening, indicating commuting behavior.
```


#### Rentals by Hour
This chart shows the average number of rentals for each hour of the day. The goal is to uncover the time blocks when rentals peak. Most users rent bikes during peak commute hours: 8 AM and 6 PM. This suggests the bikes are primarily used for work-related commuting. Knowing this allows the company to ensure bikes are restocked in high demand areas before these peak hours begin. Understanding demand flow by time can also improve predictive maintenance scheduling and customer satisfaction.


```python
# --- Visualization 2: Rentals by Temperature & Seasons ---
plt.figure(figsize=(10, 5))
sns.scatterplot(x='Temperature', y='RENTALS', data=df, hue='Season', alpha=0.7)
plt.title('Bike Rentals vs Temperature by Season')
plt.xlabel('Temperature (°C)')
plt.ylabel('Rentals')
plt.legend(title='Season')
plt.grid(True)
plt.show()

# This reveals that moderate temperatures (10-25°C) see higher rentals, especially in spring and summer.
```


#### Rentals by Temperature
This visualization examines the relationship between temperature and bike rental volume, offering insight into the conditions that drive user demand. The chart reveals a distinct pattern: rental activity increases as temperatures rise, reaching its highest levels between 20°C and 30°C. This suggests that moderate to warm weather encourages more bike usage. Conversely, demand drops when temperatures fall below 10°C or exceed 30°C, indicating that extreme cold or heat discourages ridership. These findings support the strategy of intensifying service availability and marketing efforts during favorable weather while scaling back during less ideal conditions.


```python
# --- Visualization 3: Rentals on Holidays vs Non-Holidays ---
plt.figure(figsize=(8, 5))
sns.boxplot(x='Holiday', y='RENTALS', data=df)
plt.title('Bike Rentals on Holidays vs Non-Holidays')
plt.xlabel('Holiday')
plt.ylabel('Rentals')
plt.grid(True)
plt.show()

# Users tend to rent fewer bikes on holidays, likely due to reduced commuting.
```

#### Rentals by Day Type (Holiday vs Non-Holiday)

The boxplot reveals that bike rentals are generally lower on holidays compared to regular days. This pattern suggests that most usage is tied to daily commuting, which tends to decrease when people are off work. Understanding this behavior can help with demand forecasting and operational planning around public holidays.

### Top Insights:
1. Rentals are highest in the afternoon
When we looked at the data, we saw that most people rent bikes between 12 PM and 6 PM. This time is when many people finish work or school and may want to go home, go out, or exercise. It is also a popular time for people to enjoy the outdoors. This shows that afternoon is the most important time to make sure enough bikes are available. If bikes are missing during this time, people may get upset or stop using the service.

2. Warm temperatures bring the most rentals
The data shows that people rent more bikes when the weather is warm between 20°C and 30°C. Rentals are low when it’s colder than 10°C or hotter than 30°C. This tells us that comfort matters a lot when people choose to ride a bike. On very cold or very hot days, people don’t want to ride. Knowing this, the company can prepare for busy days when the weather is just right and avoid spending too much on days with low demand.

3. Regular weekdays are busier than holidays
We also saw that regular weekdays have more rentals than holidays. Even though some holidays have busy hours, the total rentals are lower. This could be because people use bikes more for daily routines like work and school, and not just for fun. Holidays might still need some planning, especially in the afternoon, but normal weekdays are more important overall. The company should focus more on Monday through Friday and use holidays as backup planning days.

### Analysis Questions
#### **1. Why are the majority of users renting bicycles?**
The data reveals that most bike rentals occur during weekday peak hours—around 8 AM and 6 PM—which strongly aligns with traditional commuting patterns. This indicates that a significant portion of users are relying on Bay Wheels as a convenient and eco-friendly transportation option to and from work or transit stations.

Rental activity declines during midday hours and is noticeably lower on weekends and holidays, further reinforcing the conclusion that the primary driver of usage is work-related travel rather than leisure. This trend is consistent with urban mobility studies, which show that bike-sharing systems are often used to fill the "first and last mile" gap in public transport networks.

By understanding that users are primarily commuting, Bay Wheels can optimize bike availability, maintenance schedules, and station placements to better support this behavior.
#### **2. Are there ideal weather conditions that create high bike rental demand?**
Yes, the data shows that rental demand is highest in the 20°C to 30°C temperature range. Warmer temperature from 30°C to 40°C also sustain high usage, though slightly lower than the 20–30°C range. In contrast, demand is significantly lower when temperatures fall below 10°C. These results indicate that warm, comfortable weather is ideal for biking, especially in late spring and summer months. Low wind speed, good visibility, and clear skies also enhance rental activity, making weather forecasting a key operational tool.
#### **3. How should the company strategize based on your results?**
Bay Wheels should align bike availability with peak commuting hours (8 AM, 6 PM) to meet demand and optimize rebalancing efforts. Since rentals are highest in mild, dry weather, the company can run seasonal promotions in spring and summer, while offering incentives during poor conditions to sustain usage. On holidays, afternoon rentals increase, suggesting a shift from commuting to recreational use—this calls for adjusted staffing and targeted marketing. Overall, a data-driven, weather-aware, and time-sensitive strategy will boost efficiency and ridership.

This report shows that bike rentals follow some clear patterns. More people rent bikes in the afternoon and when the weather is warm, especially between 20°C and 30°C. These findings can help the company make better plans like where and when to put more bikes. By checking the weather and using this data, the company can give better service and avoid problems.
